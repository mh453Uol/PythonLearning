## 2 Cleaning Data

a. Make sure ’price’, ’bore’, ’stroke’, ’horsepower’ and ’peak-rpm’ are numeric

```python
cols = ["price", "bore", "stroke", "horsepower", "peak-rpm"]
for col in cols:
    # to_numeric("1.2",errprs="coerce") convert string number to int
    # coerce when invalid parsing will set value as NaN
    #print(data[col])
    data[col] = pd.to_numeric(data[col],errors="coerce")

# drop all rows (axis = 0)
data.dropna(axis=0,inplace=True)
data.info()
```

```python
data["lnprice"] = np.log(data.price)
```

```python
data.describe()
```

```python
def describe(df,col):
    desc = df[col].describe()
    print(desc)

describe(data,"price")
```

```python
def plotstats(df,col):
    fig = plt.figure(1)
    fig.set_size_inches(12,8)
    
    # box plot
    plt.subplot(2,1,1)
    data.dropna().boxplot(col,vert=False)
    
    #histogram
    plt.subplot(2,1,2)
    temp = data[col].as_matrix()
    plt.hist(temp, bins = 30, alpha = 0.7)
    plt.ylabel("Number of cars")
    plt.xlabel(col)
    plt.show()
    return [col]

plotstats(data,"price")
```
