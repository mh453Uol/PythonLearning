## Standard Deviasion



```python
# mean 
print("Mean fare: {0}".format(df.Fare.mean()))

# middle number
print("Median fare: {0}".format(df.Fare.median()))

# min
print("Max fare: {0}".format(df.Fare.max()))

# max
print("Min fare: {0}".format(df.Fare.min()))

# 25 % of fare
print("25 % of fare: {0}".format(df.Fare.quantile(.25)))
# 50 % of fare
print("50 % of fare: {0}".format(df.Fare.quantile(.50)))
# 75 % of fare most people paid top prices
print("75 % of fare: {0}".format(df.Fare.quantile(.75)))

#variance
print("Variance fare: {0}".format(df.Fare.var()))
# quite a huge std so people paid lots of different prices!
print("Standard deviation fare: {0}".format(df.Fare.std()))
```
